```python
plt.hist(df["data"],bins=20, color="b",alpha=0.6)
plt.xlabel("Mesi")
plt.xticks(rotation=45)
plt.ylabel("Numero totale di vaccinazioni")
plt.title("Vaccinazioni nel tempo")
plt.grid(True,linestyle="--")
plt.show()
```

diagramma a barre. scelgo asse x come preferisco

```python
plt.bar(["maschi","femmine"], [df["m"].sum(), df["f"].sum()],color=["green","pink"], label=["Maschi","Femmine"])
plt.ylabel("Numero di vaccinazioni")
plt.title("Numero delle vaccinazioni per genere")
plt.show()
```

grafico a barre nel tempo. sovrappongo maschi e femmine

```python
plt.bar(df["data"],df["m"], color="blue", alpha=0.6,label="Maschi")
plt.bar(df["data"], df["f"], color="pink", alpha=0.6, label="Femmine", bottom=df["m"])
plt.xlabel("data")
plt.ylabel("Distribuzione vaccinazioni per genere nel tempo")
plt.title("Male/Female vaccines distribution distributed over days of vaccines")
plt.show()
```

Raggruppo per mese con groupby, mantengo separate le colonne di interesse m/f

df_grouped.index.astype(str) cambia i mesi in una stringa

alpha definisce la trasparenza del colore

bottom=df_grouped crea un grafico sovrapposto

```python
df["mese_anno"] = df["data"].dt.to_period("M")
df_grouped = df.groupby("mese_anno")[["m", "f"]].sum()
plt.bar(df_grouped.index.astype(str), df_grouped["m"], color="blue", alpha=0.6, label="Maschi")
plt.bar(df_grouped.index.astype(str), df_grouped["f"], color="pink", alpha=0.6, label="Femmine", bottom=df_grouped["m"])
plt.title("Male/Female vaccines distribution distributed over months")
plt.show()
```

```python
plt.pie([df["m"].sum(),df["f"].sum()], labels=["Maschi","femmine"],autopct='%.2f%%')
plt.title("Distribuzione vaccinazioni per genere")
plt.show()
```
