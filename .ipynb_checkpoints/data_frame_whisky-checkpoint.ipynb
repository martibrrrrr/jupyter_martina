```python
import requests
import pandas as pd
import matplotlib.pylab as plt
```

```python
url = "https://whiskyhunter.net/api/auctions_data"
response = requests.get(url)
```

```python
data = response.json()
data?
```

```python
print(data[0])
```

```python
df = pd.DataFrame(data)
print(df.head())
```

Cerco la bottiglia più costosa (Nome)

```python
df['dt'] = pd.to_datetime(df['dt'])
```

```python
df_price=df["winning_bid_max"]
```

```python
highest_price = df.loc[df_price.idxmax()]
```

```python
print(f"La bottiglia più costosa è stata venduta il {highest_price['dt']} a {highest_price['winning_bid_max']}€")
```

# Andamento dei prezzi delle bottiglie di whisky all'asta

```python
plt.plot(df['dt'], df['winning_bid_max'], marker='o', linestyle='-', label="Prezzo max")
plt.plot(df['dt'], df['winning_bid_mean'], marker='s', linestyle='--', label="Prezzo medio")
plt.plot(df['dt'], df['winning_bid_min'], marker='o', linestyle='-', label="Prezzo min")
plt.xlabel("Data")
plt.ylabel("Prezzo (€)")
plt.title("Andamento dei prezzi delle bottiglie di whisky all'asta")
plt.legend()
plt.show()
```

# Distribuzione dei prezzi massimi delle bottiglie all'asta

```python
df_sorted = df.sort_values(by="winning_bid_max", ascending=False)
```

```python
print(df_sorted[['dt', 'auction_name', 'winning_bid_max']].head(10))
```

```python
plt.scatter(df['dt'], df['winning_bid_max'])
plt.xlabel("Data")
plt.ylabel("Prezzo massimo (€)")
plt.title("Distribuzione dei prezzi massimi delle bottiglie all'asta")
plt.show()
```

# Prezzo medio delle bottiglie per asta

```python
df_grouped = df.groupby("auction_name")["winning_bid_mean"].mean().sort_values(ascending=False)
df_grouped.plot(kind='bar', color='orange')
plt.xlabel("Nome dell'asta")
plt.ylabel("Prezzo medio (€)")
plt.title("Prezzo medio delle bottiglie per asta")
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

# Distribuzione dei lotti tra le diverse aste

```python
df_lots = df.groupby("auction_name")["auction_lots_count"].sum()
df_lots = df_lots.sort_values()
df_lots.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.ylabel("")
plt.title("Distribuzione dei lotti tra le diverse aste")
plt.show()
```
