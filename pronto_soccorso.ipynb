```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://dati.toscana.it/dataset/e642762c-a7da-433c-a18c-3148bf50c557/resource/624fbe40-8b7e-4721-bf28-a4f4c314c411/download/accessiprontosoccorso2023.csv"
df = pd.read_csv(url,sep=";")
print(df.head())
```

# Numero di accessi in ospedale per livello di gravità

```python
plt.bar(df["GRAVITA_ACCETTAZIONE "], df["ACCESSI_TOTALI"], color="skyblue")
plt.xlabel("Gravità Accettazione")
plt.ylabel("Numero di Accessi")
plt.title("Numero di accessi per livello di gravità")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

# Accessi suddivisi per tipologia di Pronto Soccorso nei vari livelli di gravità

```python
gravitity_count = df.groupby("GRAVITA_ACCETTAZIONE ")[["ACCESSI_PS_GENERALE", "ACCESSI_PEDIATRICI", "ACCESSI_OCULISTICI"]].sum()
```

```python
gravitity_count.plot(kind='bar', figsize=(12, 8), color=["skyblue", "lightgreen", "salmon"])
plt.xlabel("Gravità Accettazione")
plt.ylabel("Numero di Accessi")
plt.title("Accessi per tipologia di Pronto Soccorso nei vari livelli di gravità")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.legend(["Generale", "Pediatrico", "Oculistico"], loc="upper left")
plt.show()
```

# Percentuale di accessi per tipologia di Pronto Soccorso

```python
df_totali = df[["ACCESSI_PS_GENERALE", "ACCESSI_PEDIATRICI", "ACCESSI_OCULISTICI"]].sum()

plt.figure(figsize=(8, 8))
df_totali.plot(kind="pie", autopct="%1.1f%%", cmap="viridis", startangle=90)
plt.ylabel("")
plt.title("Distribuzione degli accessi per tipologia di Pronto Soccorso")
plt.show()
```
